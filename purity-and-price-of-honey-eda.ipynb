```python
def add_random_missing_values(dataframe: pd.DataFrame,
                              missing_rate: float = 0.05,
                              random_state: int = 42,
                              exclude_columns: list = None) -> pd.DataFrame:
    """Turns random values to NaN in a DataFrame with reproducibility, excluding specified columns.

    Args:
        dataframe (pd.DataFrame): DataFrame to be processed.
        missing_rate (float): Percentage of missing value rate in float format. Defaults 0.05.
        random_state (int): Seed for random number generator. Defaults 42.
        exclude_columns (list): List of column names to exclude from missing value assignment.

    Returns:
        df_missing (pd.DataFrame): Processed DataFrame object.

    """
    # Set random seed for reproducibility
    np.random.seed(random_state)

    # Get copy of dataframe
    df_missing = dataframe.copy()

    # Determine columns to include
    if exclude_columns:
        columns_to_include = [col for col in dataframe.columns if col not in exclude_columns]
    else:
        columns_to_include = dataframe.columns.tolist()

    # Ensure we have at least one column to modify
    if not columns_to_include:
        raise ValueError("All columns are excluded. No missing values can be added.")

    # Obtain size of dataframe and number of total missing values
    df_size = len(columns_to_include) * dataframe.shape[0]
    num_missing = int(df_size * missing_rate)

    # Map included column indices
    included_col_indices = [dataframe.columns.get_loc(col) for col in columns_to_include]

    # Generate random row and column indexes
    row_indices = np.random.randint(0, dataframe.shape[0], num_missing)
    col_indices = np.random.choice(included_col_indices, num_missing)

    # Turn selected values into NaN
    for row_idx, col_idx in zip(row_indices, col_indices):
        df_missing.iat[row_idx, col_idx] = np.nan

    return df_missing


# Veri çerçevesine %0.05 NaN ekle
df = add_random_missing_values(data, missing_rate=0.005,random_state=42,
                               exclude_columns=["Pollen_analysis","Price"])
df.head() # Veri setinin ilk 5 satırını getirir
```

```python
# Summary Function
def get_summary(df: pd.DataFrame) -> pd.DataFrame:
    """
    Generates a summary of the given DataFrame, including descriptive statistics and additional information.

    This function returns a summary DataFrame with the following columns for each feature:
    - Count: Non-missing entries in each column.
    - unique: Number of unique values in each column.
    - missing: Number of missing values in each column.
    - duplicated: Total number of duplicate rows in the DataFrame.
    - mean: Mean of the column (for numerical columns only).
    - std: Standard deviation of the column (for numerical columns only).
    - min: Minimum value in the column (for numerical columns only).
    - 25%: 25th percentile of the column (for numerical columns only).
    - 50%: 50th percentile (median) of the column (for numerical columns only).
    - 75%: 75th percentile of the column (for numerical columns only).
    - max: Maximum value in the column (for numerical columns only).
    
    Args:
        df (pd.DataFrame): The input DataFrame to be summarized.

    Returns:
        pd.DataFrame: A styled DataFrame with descriptive statistics and additional information,
                      including background color gradients and borders for visual emphasis.
    """
    df_desc = pd.DataFrame(df.describe(include="all").T)
    df_summary = pd.DataFrame({"dtype": df.dtypes,
                               "Count": df.count(),
                               "unique": df.nunique(),
                               "missing": df.isna().sum(),
                               "duplicated": df.duplicated().sum(),
                               "mean": df_desc["mean"].values,
                               "min": df_desc["min"].values,
                               "std": df_desc["std"].values,
                               "25%": df_desc["25%"].values,
                               "50%": df_desc["50%"].values,
                               "75%": df_desc["75%"].values,
                               "max": df_desc["max"].values})

    return df_summary.style\
        .background_gradient(cmap='YlGnBu', subset=['mean', 'std', 'min', '25%', '50%', '75%', 'max'])\
        .set_properties(**{'border': '1.5px solid black'})
```

# Descriptive Analysis

## T-Test

#### 1. **Değişkenler Arası Farkları İncelemek:**

* Bu analiz, iki sürekli değişken arasındaki farkların istatistiksel anlamlılık taşıyıp taşımadığını belirler. Kod, veri setindeki tüm sürekli değişkenler için çiftler arasındaki t-testlerini gerçekleştirir. Bu testler, her iki değişkenin ortalamaları arasındaki farkın tesadüfi olup olmadığını test eder. Eğer p-değeri küçükse (örneğin, 0.05'ten küçükse), değişkenler arasında anlamlı bir fark olduğu sonucuna varılır.
* Kodda ayrıca Bonferroni düzeltmesi uygulanmaktadır. Birden fazla t-testi yapıldığında, yanlış pozitif sonuçlar (Type I error) riskinin arttığı bilinir. Bu riskin önüne geçmek için Bonferroni düzeltmesi kullanılır. Bu düzeltme, her bir test için p-değerini test sayısı ile çarparak, çoklu testlerin neden olabileceği hata olasılığını azaltır.


#### 2. **Çoklu Testlerin Düzeltilmesi - Bonferroni Düzeltmesi:**
* Kodda ayrıca Bonferroni düzeltmesi uygulanmaktadır. Birden fazla t-testi yapıldığında, yanlış pozitif sonuçlar (Type I error) riskinin arttığı bilinir. Bu riskin önüne geçmek için Bonferroni düzeltmesi kullanılır. Bu düzeltme, her bir test için p-değerini test sayısı ile çarparak, çoklu testlerin neden olabileceği hata olasılığını azaltır.

#### 3. **Anlamlılık Testi Sonuçlarının Görselleştirilmesi:**

- **t-statistic:** Bu değer, her iki değişkenin ortalamaları arasındaki farkın büyüklüğünü ve bu farkın örneklemden ne kadar uzak olduğunu gösterir. Yüksek t-statistik değeri, iki değişken arasındaki farkın büyük olduğunu ve dolayısıyla anlamlı olma olasılığının yüksek olduğunu gösterir.
- **p-value:** p-değeri, testin sonucunun tesadüfi olup olmadığını gösterir. Eğer p-değeri 0.05'ten küçükse, bu değişkenler arasında anlamlı bir fark olduğu kabul edilir.
- **Bonferroni adjusted p-value:** Bonferroni düzeltmesiyle düzeltilmiş p-değeri, çoklu testlerden kaynaklanan hata riskini azaltmak için her p-değerini test sayısıyla çarpar.
- **Bonferroni significant:** Eğer düzeltilmiş p-değeri 0.05'ten küçükse, bu durum iki değişken arasında anlamlı bir fark olduğunu gösterir.

#### Bu analiz neden yapılır?

Bu tür analizler genellikle aşağıdaki durumlar için kullanılır:

- **Değişkenler arasındaki ilişkiyi anlamak:** Bir değişkenin diğerleriyle olan ilişkisini incelemek.
- **Modelleme öncesi analiz:** Özellikle regresyon ya da sınıflandırma modellemesi yapmadan önce, değişkenlerin birbirleriyle ilişkilerini anlamak ve hangi değişkenlerin anlamlı olduğunu belirlemek.
- **Veri keşfi:** Veri setindeki farklı özellikler arasındaki ilişkileri daha iyi anlamak.

```python
# Tüm sürekli değişkenlerin isimlerini al
features = numerical_columns

# Sonuçları tutmak için bir liste
t_test_results = []

# Çiftler arasında t testi uygula
for i in range(len(features)):
    for j in range(i + 1, len(features)):
        var1 = features[i]
        var2 = features[j]
        t_stat, p_value = ttest_ind(df[var1], df[var2],nan_policy="omit")
        t_test_results.append({
            'Variable 1': var1,
            'Variable 2': var2,
            't-statistic': t_stat,
            'p-value': p_value
        })

# Sonuçları bir DataFrame'e çevir
results_df = pd.DataFrame(t_test_results)


# Bonferroni Düzeltmesi
num_tests = len(results_df)  # Toplam test sayısı
alpha = 0.05  # Varsayılan anlamlılık seviyesi
results_df['Bonferroni adjusted p-value'] = results_df['p-value'] * num_tests
results_df['Bonferroni adjusted p-value'] = results_df['Bonferroni adjusted p-value'].clip(upper=1)
results_df['Bonferroni significant'] = results_df['Bonferroni adjusted p-value'] < alpha


# Sonuçları yazdır

display(results_df.sort_values("t-statistic",ascending=False)[:10])
```

### pH ve Purity için ANOVA testi
Eğer pH ve saflık arasında fark olup olmadığını test etmek istiyorsanız, ANOVA (Analysis of Variance) veya t-test gibi istatistiksel testler yapabilirsiniz.

Örneğin, pH değeri farklı olan gruplar arasında saflık açısından anlamlı bir fark olup olmadığını görmek için ANOVA testi kullanabilirsiniz:

```python
from scipy.stats import f_oneway

# pH gruplarını ayır
low_pH = group_df[group_df['pH_group'] == 'Low']['Purity']
medium_pH = group_df[group_df['pH_group'] == 'Medium']['Purity']
high_pH = group_df[group_df['pH_group'] == 'Normal']['Purity']

# Eksik verileri temizle
low_pH_clean = low_pH.dropna()
medium_pH_clean = medium_pH.dropna()
high_pH_clean = high_pH.dropna()

# ANOVA testi
F_statistic, p_value = f_oneway(low_pH_clean, medium_pH_clean, high_pH_clean)
print(f"ANOVA Sonucu: F = {F_statistic} p = {p_value:.4f}")
```

p_value < 0.05 olduğundan pH değeri farklı olan gruplar arasında saflık açısından anlamlı bir fark vardır.

## Metrics for Numerical Features Based on Robustness to Outliers

Farklı metriklerin aykırı değerleri nasıl ele aldığını anlamak veri analizinde çok önemlidir. Aşağıda, aykırı değerlere karşı sağlamlıklarına göre kategorize edilmiş sayısal özellikler için temel metrikler yer almaktadır.

### Outliers
Aykırı değerler, veri kümesindeki diğer gözlemlerden önemli ölçüde farklı olan veri noktalarıdır. Bunlar verilerdeki değişkenliğin bir sonucu olabilir veya ölçüm hatasına işaret edebilir. Aykırı değerlerin belirlenmesi ve ele alınması çok önemlidir çünkü istatistiksel analizleri büyük ölçüde etkileyebilirler.

---
- **Robust to Outliers**                  
    - Median
    - Interquartile Range (IQR)
    - Median Absolute Deviation (MAD)
    - Trimmed Mean
    
- **Sensitive to Outliers**                    
    - Mean
    - Standard Deviation
    - Variance



### Skewness and Kurtosis
**Çarpıklık(Skewness)**, bir dağılımın asimetrisinin bir ölçüsüdür. Bir dağılım, sol ve sağ tarafları ayna görüntüsü olmadığında asimetriktir. Bir dağılım sağ (veya pozitif), sol (veya negatif) veya sıfır çarpıklığa sahip olabilir. [Detay](https://tr.wikipedia.org/wiki/%C3%87arp%C4%B1kl%C4%B1k)

* Symmetric Skewness for Normal Distribution (=0) 
* Positive (Right) Skewness (>0)
* Negative (left) Skewness (<0)
  
![](https://datakuity.com/wp-content/uploads/2021/08/sk1-1024x386.png)

**Basıklık(Kurtosis)**, bir dağılımın kuyrukluluğunun bir ölçüsüdür. Kuyrukluluk, aykırı değerlerin ne sıklıkta ortaya çıktığıdır. Aşırı basıklık, bir dağılımın normal dağılıma göre kuyruklu olmasıdır. Orta basıklığa (orta kuyruk) sahip dağılımlar mezokurtiktir. Düşük basıklığa sahip (ince kuyruklu) dağılımlar platikurtiktir. [Detay](https://en.wikipedia.org/wiki/Kurtosis)

* Mesokurtic (Normal Distribution) (=3) 
* Leptokurtic (>3)
* Platykurtic (<3)

![](https://keytodatascience.com/wp-content/uploads/2021/11/Kurtosis1.jpg)

```python
# Function to calculate metrics for skewed and normal distributions
def calculate_metrics(column_data):
    # Basic statistics
    mean_value = column_data.mean()
    median_value = column_data.median()
    std_dev_value = column_data.std()
    variance_value = column_data.var()
    confidence_interval = stats.norm.interval(0.95, loc=mean_value, scale=stats.sem(column_data))
    
    # Advanced statistics
    skewness_value = column_data.skew()
    kurtosis_value = column_data.kurtosis()
    mad_value = stats.median_abs_deviation(column_data)
    iqr_value = stats.iqr(column_data)

    return {
        "Mean": mean_value,
        "Median": median_value,
        "Standard Deviation": std_dev_value,
        "Variance": variance_value,
        "95% Confidence Interval": confidence_interval,
        "Skewness": skewness_value,
        "Kurtosis": kurtosis_value,
        "MAD": mad_value,
        "IQR": iqr_value
    }

# Iterate over each continuous column to calculate and print statistics
for col in numerical_columns:
    col_data = fill_group_df[col]
    metrics = calculate_metrics(col_data)
    
    # Print the statistics with explanations
    print(f"\nStatistics for {col}:")
    print(f"{'Metric':<35} {'Value':<40} ")
    print(f"{'-'*35} {'-'*40}")
    print(f"{'Mean':<35} {metrics['Mean']:<40} ")
    print(f"{'Median':<35} {metrics['Median']:<40} ")
    print(f"{'Standard Deviation':<35} {metrics['Standard Deviation']:<40} ")
    print(f"{'Variance':<35} {metrics['Variance']:<40} ")
    print(f"{'95% Confidence Interval':<35} {str(metrics['95% Confidence Interval']):<40} ")
    print(f"{'Skewness':<35} {metrics['Skewness']:<40} ")
    print(f"{'Kurtosis':<35} {metrics['Kurtosis']:<40} ")
    print(f"{'MAD':<35} {metrics['MAD']:<40} ")
    print(f"{'IQR':<35} {metrics['IQR']:<40} \n")
```

### Outliers

```python
def drop_outlier(df: pd.DataFrame, columns_list: list) -> pd.DataFrame:
    """
    Removes rows containing outliers in specified columns from the DataFrame.

    Parameters:
    df (pd.DataFrame): Input DataFrame.
    columns_list (list): List of column names to check for outliers.

    Returns:
    pd.DataFrame: DataFrame with outliers removed.
    """
    data = df.copy()
    print("Old data shape: ",data.shape) 
    
    for col in columns_list:
        q1 = data[col].quantile(0.25)
        q3 = data[col].quantile(0.75)
        IQR = q3 - q1
        fence_low = q1 - (1.5 * IQR)
        fence_high = q3 + (1.5 * IQR)
        
        # Filter rows where column values are within the fence
        data = data[(data[col] >= fence_low) & (data[col] <= fence_high)]

    print("New data shape: ",data.shape) 
    
    return data
```

# Feature Engineering

## Feature Screening

Özellik taraması, veri kalitesi sürecinde analize veya modellemeye anlamlı bilgi katmayan özelliklerin (değişkenlerin) belirlenmesini ve kaldırılmasını içeren çok önemli bir adımdır. Bu tür özellikleri eleyerek veri setini düzene sokabilir, model performansını artırabilir ve yorumlanabilirliği geliştirebiliriz. Bu bölümde, özellik taraması için üç özel kriteri ele alacağız:

**Sürekli Değişkenler için 0,1'den Küçük Varyasyon Katsayısına Sahip Özellikler**

Varyasyon katsayısı (CV) göreceli değişkenliğin bir ölçüsüdür. Standart sapmanın ortalamaya oranı olarak hesaplanır. CV değeri 0,1'den düşük olan özellikler düşük değişkenliğe sahip olarak kabul edilir ve analiz için önemli bilgiler sağlamayabilir. Bu tür özellikleri belirleyecek ve kaldıracağız.

**Kategorik Değişkenler için Mod Kategorisi Yüzdesinin %95'ten Büyük Olduğu Özellikler**

Tek bir kategorinin baskın olduğu kategorik değişkenler (mod kategori yüzdesi > %95) fazla varyasyon sağlamadıkları için analiz için kullanışlı olmayabilir. Veri setini düzene sokmak için bu kategorik özellikleri belirleyip kaldıracağız.

**Kategorik Değişkenler için Benzersiz Kategori Yüzdesi %90'ı Aşan Özellikler**

Yüksek oranda benzersiz kategoriye sahip kategorik değişkenler (>%90) analizi karmaşıklaştırabilir ve modellerde aşırı uyuma yol açabilir. Daha sağlam ve genelleştirilebilir bir model sağlamak için bu özellikleri belirleyecek ve kaldıracağız.

Bu eleme kriterlerini uygulayarak, veri kümesinde kalan özelliklerin sonraki analizler için anlamlı ve ilgili bilgiler sağlamasını temin edebiliriz.


```python
def feature_screening(df: pd.DataFrame,
                      target: str, 
                      categorical_columns: list, 
                      numerical_columns: list) -> pd.DataFrame:
    """
    Performs feature screening on a dataset by analyzing numerical and categorical features 
    to identify and remove less informative or redundant features.

    Parameters:
    -----------
    df : pd.DataFrame
        The dataset containing features and the target variable.

    target : str
        The name of the target variable column.

    categorical_columns : list
        A list of column names corresponding to categorical features.

    numerical_columns : list
        A list of column names corresponding to numerical features.

    Returns:
    --------
    pd.DataFrame
        A cleaned dataset with the identified features removed.

    Methodology:
    ------------
    - For numerical features:
      - Calculates the Coefficient of Variation (CV) and identifies features with CV < 0.1.

    - For categorical features:
      - Identifies features where the most frequent category (mode) exceeds 95% of the data.
      - Identifies features where the percentage of unique categories exceeds 90%.

    - Combines all identified features to be removed and drops them from the dataset.

    Example:
    --------
    >>> import pandas as pd
    >>> data = {
    ...     'num1': [1, 1, 1, 1],
    ...     'num2': [2, 2, 2, 2],
    ...     'cat1': ['A', 'A', 'A', 'A'],
    ...     'cat2': ['B', 'C', 'D', 'E'],
    ...     'target': [0, 1, 0, 1]
    ... }
    >>> df = pd.DataFrame(data)
    >>> cleaned_data = feature_screening(
    ...     df=df, 
    ...     target='target', 
    ...     categorical_columns=['cat1', 'cat2'], 
    ...     numerical_columns=['num1', 'num2']
    ... )
    >>> print(cleaned_data)
    """
    # Separate the dataset into input variables (predictors) and target variable (response)
    label = df[target]
    inputs = df.drop(columns=[target])
    
    # Calculate Coefficient of Variation for continuous variables
    cv = inputs[numerical_columns[:-1]].std() / inputs[numerical_columns[:-1]].mean()
    
    # Identify features with CV less than 0.1
    low_cv_features = cv[cv < 0.1].index.tolist()
    print("Features with Coefficient of Variation less than 0.1:", low_cv_features)
    
    # Calculate Mode Category Percentage for categorical variables
    mode_percentage = inputs[categorical_columns].apply(lambda x: x.value_counts(normalize=True).max() * 100)
    
    # Identify features where the mode category percentage is greater than 95%
    high_mode_features = mode_percentage[mode_percentage > 95].index.tolist()
    print("Categorical features where mode category percentage is greater than 95%:", high_mode_features)
    
    # Calculate Percentage of Unique Categories for categorical variables
    unique_category_percentage = inputs[categorical_columns].nunique() / len(inputs) * 100
    
    # Identify features with a percentage of unique categories exceeding 90%
    high_unique_features = unique_category_percentage[unique_category_percentage > 90].index.tolist()
    print("Categorical features with percentage of unique categories exceeding 90%:", high_unique_features)
    
    # Combine all features to be removed
    features_to_remove = set(low_cv_features + high_mode_features + high_unique_features)
    print("Features to be removed:", features_to_remove)
    
    # Remove the identified features from the inputs dataframe
    cleaned_inputs = inputs.drop(columns=features_to_remove)
    
    # Combine the cleaned inputs with the label
    cleaned_dataset = pd.concat([cleaned_inputs, label], axis=1)
    
    # Display the cleaned dataset
    return cleaned_dataset
```
